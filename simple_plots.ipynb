# Reviewing for the Unit Test: Plotting Functions

### Instructions:

Plot the following functions for $-\pi \leq x \leq \pi$:

- $y_0 = \sin{x}$ (*use a solid blue line*)

- $y_1 = x$ (*use a dotted red line*)

- $y_2 = x - \frac{x^3}{6}$ (*use a dotted green line*)

- $y_3 = x - \frac{x^3}{6} + \frac{x^5}{125}$ (*use a dotted blue line*)

### Commentary:

We can take two approaches to this problem - one using list comprehensions in "pure" Python, and the other taking advantage of some of `numpy`'s built in goodness. The `numpy` version results in easier to read code, so we'll start there.


```python
import numpy as np
import matplotlib.pyplot as plt

# Not strictly necessary, but allows us to change constants in one place
# Avoids "magic" values
x_min = -np.pi
x_max = np.pi
n = 100

# Use Numpy's linspace function to generate n evenly spaced points
x_values = np.linspace(x_min, x_max, n)

# Evaluating the y values is as simple as performing mathematical operations
# on the arrays as a whole since these operations are *vectorized* in Numpy
y0 = np.sin(x_values)
y1 = x_values
y2 = y1 - x_values ** 3 / 6
y3 = y2 + x_values ** 5 / 125

# Straightforward matplotlib stuff. Here we opted to use the format strings
# for color and linestyle instead of explicitly using keyword arguments.
# The details of these formats can be found in the documentation here:
# https://matplotlib.org/3.5.1/api/_as_gen/matplotlib.pyplot.plot.html#:~:text=Notes-,Format%20Strings,-A%20format%20string

plt.plot(x_values, y0, '-b', label="$\sin{x}$")
plt.plot(x_values, y1, '--r', label="x")
plt.plot(x_values, y2, '--g', label="x - x^3 / 6")
plt.plot(x_values, y3, '--b', label="x - x^3 / 6 + x^5 / 125")
plt.legend()
plt.grid()
plt.show()
```

In the so-called "pure" approach, we'll need to use list comprehensions to generate lists of x and y values instead of operating on `array` objects. Since Python's `range()` function only allows us to use integer arguments, we'll have to get creative with generating our x values.

```python
import math
import matplotlib.pyplot as plt

# Same notes as above here
x_min = -np.pi
x_max = np.pi
n = 100

# Remember that n points divide a line into (n - 1) equal pieces!
x_values = [x_min + i * (x_max - x_min) / (n - 1) for i in range(n)]

# Here we use listcomps to apply each formula to each x value
y0 = [math.sin(x) for x in x_values]
y1 = [x for x in x_values] # or just x_values
y2 = [x - x**3 / 6 for x in x_values]
y3 = [x - x**3 / 6 + x**5 / 125 for x in x_values]

# Same plot code as above
plt.plot(x_values, y0, '-b', label="$\sin{x}$")
plt.plot(x_values, y1, '--r', label="x")
plt.plot(x_values, y2, '--g', label="x - x^3 / 6")
plt.plot(x_values, y3, '--b', label="x - x^3 / 6 + x^5 / 125")
plt.legend()
plt.grid()
plt.show()
```
